```python
from google_play_scraper import app
import pandas as pd
import numpy as np
from google_play_scraper import Sort, reviews_all,reviews
```

```python
appnames = [
    'com.black_crypto',
    'com.king.candycrushsaga',
    'com.dts.freefireth',
    'com.instagram.android',
    'com.snapchat.android',
    'com.mxtech.videoplayer.ad',
    'com.flipkart.android',
    'com.pubg.imobile',
    'com.kiloo.subwaysurf',
    'com.ludo.king',
    'org.telegram.messenger',
    'com.phonepe.app',
    'com.truecaller',
    'com.facebook.katana',
    'net.one97.paytm',
    'com.whatsapp',
    'com.adobe.lrmobile',
    'com.supercell.clashofclans',
    'com.activision.callofduty.shooter',
    'com.miniclip.eightballpool',
]
```

```python
df_list = []
for app in appnames:
    review = reviews(
        app,
        lang='en', # defaults to 'en'
        country='in', # defaults to 'us'
        sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
        count=1000
    )
    for row in review:
        try:
            for record in row:
                record['app']=app
            df_list.extend(row)
        except Exception as e:
            print(e)
print(len(df_list))
df = pd.DataFrame(df_list)
df.shape
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///db.sqlite3',)
```

```python
from textblob import TextBlob
```

```python
df.columns.tolist()
```

```python
def sentiment_analysis(review):
  def getSubjectivity(text):
    try:
      return TextBlob(text).sentiment.subjectivity
    except:
      return None
  
  #Create a function to get the polarity
  def getPolarity(text):
    try:
      return TextBlob(text).sentiment.polarity
    except:
      return None

  def getAnalysis(score):
    if isinstance(score,(int,float)):
      if score < - 0.5:
        return 'Very Negative'
      elif score >= -0.5 and score < 0:
        return "Negative"
      elif score == 0:
        return 'Neutral'
      elif score > 0 and score <=0.5:
        return 'Positive'
      else:
        return 'Very Positive'
  
  #Create two new columns ‚ÄòSubjectivity‚Äô & ‚ÄòPolarity‚Äô
  try:
    review['subjectivity'] =   review['content'].apply(getSubjectivity)
    review ['polarity'] = review['content'].apply(getPolarity)
    review ['analysis'] = review['polarity'].apply(getAnalysis)
  except Exception as e:
    print(e)
  return review
```

```python
result = sentiment_analysis(df)
```
